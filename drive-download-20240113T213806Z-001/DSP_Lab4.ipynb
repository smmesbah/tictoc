```python
import numpy as np
import matplotlib.pyplot as plt
import random

x=[]
L=300
for i in range(L):
  n=random.randint(1,10)
  x.append(n)

x=np.divide(x,10)
plt.plot(x)
```

```python
#moving average
y=[]
for i in range(1,L-1):
  y.append((x[i-1]+x[i]+x[i+1])/3)

plt.plot(y)
```

```python
#accumulator
acc=[]
for k in range(L):
  acc.append(sum(x[0:k+1]))

plt.plot(acc)
```

```python
#1st order difference
fod=[]

for k in range(1,L-1):
  fod.append(x[k]-x[k-1])

plt.plot(fod)
```

```python
#median
med=[]
for i in range(2,L-2):
  med.append(sorted([x[i-2],x[i-1],x[i],x[i+1],x[i+2]])[2])

plt.plot(med)
```
