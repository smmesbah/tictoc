```python
import numpy as np
import matplotlib.pyplot as plt

f=10
Fs=200
L=65
x=np.zeros(L)
for n in range(L):
  x[n]=np.cos(2*np.pi*f*n/Fs)

plt.stem(x)
```

```python
#Unit impulse
d=np.zeros(L)
d[L//2]=1
x=np.arange(-L//2,L//2)
plt.stem(x,d)
```

```python
#Unit Ramp
d=np.zeros(L)
x=np.arange(-L//2,L//2)
for n in range(L//2,L):
  if(n>0):
    d[n]=n-L//2

plt.stem(x,d)
```

```python
#Exponential
d=np.zeros(L)
a=0.95
for n in range(0,L):
  d[n]=a**n

plt.stem(d)
```
