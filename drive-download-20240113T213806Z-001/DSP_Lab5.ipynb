```python
import numpy as np

# Define the two discrete time signals
x = np.array([1, 2, 3, 4, 5])
h = np.array([0.5, 1, 0.5])

# Convolve the two signals
y = np.convolve(x, h, mode='full')

# Print the convolution result
print(y)
```

```python
import matplotlib.pyplot as plt

# Plot the signals
plt.figure(figsize=(10, 3))
plt.subplot(131)
plt.stem(x, use_line_collection=True)
plt.title('x[n]')
plt.subplot(132)
plt.stem(h, use_line_collection=True)
plt.title('h[n]')
plt.subplot(133)
plt.stem(y, use_line_collection=True)
plt.title('y[n]')
plt.tight_layout()
plt.show()
```
